```python
# ============================================================
# CELDA 3: Encontrar el modelo best.pt en Drive
# ============================================================


print("🔍 Buscando el archivo best.pt en tu Drive...")
print("=" * 60)

drive_path = '/content/drive/MyDrive'
modelos_encontrados = []

for root, dirs, files in os.walk(drive_path):
    for file in files:
        if file == 'best.pt':
            full_path = os.path.join(root, file)
            size_mb = os.path.getsize(full_path) / 1024 / 1024
            modelos_encontrados.append((full_path, size_mb))
            print(f"📦 Encontrado: {full_path}")
            print(f"   Tamaño: {size_mb:.2f} MB")
            print()

if not modelos_encontrados:
    print("⚠️ No se encontró ningún best.pt en tu Drive.")
    print("   Verifica que el archivo exista en la carpeta del proyecto.")
else:
    print(f"✅ Total de modelos encontrados: {len(modelos_encontrados)}")
    print("\n💡 Usaremos el primero (o el más grande si hay varios)")
```

```python
# ============================================================
# CELDA 4: Instalar Ultralytics y cargar el modelo
# ============================================================



# ============================================================
# Cargar el modelo
# ============================================================

# Ruta del modelo en Drive
MODEL_DRIVE_PATH = '/content/drive/MyDrive/PDI_Tumores_Cerebrales/best.pt'

# Copiamos el modelo a /content para trabajarlo localmente (más rápido)
MODEL_LOCAL_PATH = '/content/best.pt'
shutil.copy(MODEL_DRIVE_PATH, MODEL_LOCAL_PATH)
print(f"📦 Modelo copiado a: {MODEL_LOCAL_PATH}")

# Cargar el modelo
print("\n🔄 Cargando modelo YOLOv8...")
model = YOLO(MODEL_LOCAL_PATH)
print("✅ Modelo cargado exitosamente\n")

# ============================================================
# Información del modelo
# ============================================================
print("=" * 60)
print("📊 INFORMACIÓN DEL MODELO")
print("=" * 60)
print(f"\n🏷️ Nombre del modelo: YOLOv8n (nano)")
print(f"📁 Ruta: {MODEL_LOCAL_PATH}")
print(f"\n🎯 Clases que detecta ({len(model.names)} clases):")
for idx, name in model.names.items():
    print(f"   {idx}: {name}")

print(f"\n⚙️ Información técnica:")
print(f"   - Tarea: {model.task}")
print(f"   - Dispositivo actual: {model.device}")

```

```python
# ============================================================
# CELDA 5: Exportar el modelo a TorchScript
#
# TorchScript es el formato intermedio de PyTorch.
# Es necesario como paso previo para muchas optimizaciones,
# incluyendo ExecuTorch.
# ============================================================

print("=" * 60)
print("🔧 EXPORTACIÓN A TORCHSCRIPT")
print("=" * 60)

try:
    print("\n🔄 Exportando a TorchScript...")
    torchscript_path = model.export(
        format='torchscript',
        imgsz=640,
        optimize=True
    )
    print(f"\n✅ Exportación exitosa!")
    print(f"📁 Modelo TorchScript guardado en: {torchscript_path}")

    # Verificar el archivo
    import os
    size_mb = os.path.getsize(torchscript_path) / 1024 / 1024
    print(f"📊 Tamaño del archivo: {size_mb:.2f} MB")

except Exception as e:
    print(f"\n❌ Error en la exportación: {e}")
    print("Intentaremos con un formato alternativo")
```

```python
# ============================================================
# CELDA 6: Exportar el modelo a ONNX
#
# ONNX (Open Neural Network Exchange) es un formato estándar
# de la industria para modelos de IA. Permite ejecución
# multi-plataforma y es ampliamente usado en producción.
# ============================================================

print("=" * 60)
print("🔧 EXPORTACIÓN A ONNX")
print("=" * 60)

try:
    print("\n🔄 Exportando a ONNX...")
    onnx_path = model.export(
        format='onnx',
        imgsz=640,
        opset=12,
        simplify=True,
        dynamic=False
    )
    print(f"\n✅ Exportación exitosa!")
    print(f"📁 Modelo ONNX guardado en: {onnx_path}")

    # Verificar el archivo
    import os
    size_mb = os.path.getsize(onnx_path) / 1024 / 1024
    print(f"📊 Tamaño del archivo: {size_mb:.2f} MB")

except Exception as e:
    print(f"\n❌ Error en la exportación: {e}")
```

```python
# ============================================================
# CELDA 7: Exportación a ExecuTorch
#
# ExecuTorch es el framework de PyTorch para desplegar modelos
# en dispositivos edge (móviles, embebidos, IoT).
#
# Nota: ExecuTorch es un framework relativamente nuevo (2024)
# y su compatibilidad con modelos complejos como YOLOv8 puede
# presentar retos técnicos que documentaremos.
# ============================================================


print("=" * 60)
print("🔧 EXPORTACIÓN A EXECUTORCH")
print("=" * 60)

# ============================================================
# PASO 1: Instalar ExecuTorch
# ============================================================
print("\n📦 Paso 1: Instalando ExecuTorch...")
print("⏱️ Esto puede tomar 2-5 minutos...\n")

try:
    # Instalar executorch (la versión más estable)
    result = subprocess.run(
        [sys.executable, "-m", "pip", "install", "executorch", "-q"],
        capture_output=True,
        text=True,
        timeout=600
    )

    if result.returncode == 0:
        print("✅ ExecuTorch instalado exitosamente")
        executorch_disponible = True
    else:
        print(f"⚠️ Error en instalación:\n{result.stderr[-500:]}")
        executorch_disponible = False

except subprocess.TimeoutExpired:
    print("⏰ Timeout en instalación (>10 min)")
    executorch_disponible = False
except Exception as e:
    print(f"❌ Error inesperado: {e}")
    executorch_disponible = False

# ============================================================
# PASO 2: Intentar exportación
# ============================================================
print("\n" + "=" * 60)
print("📦 Paso 2: Intentar exportación a ExecuTorch")
print("=" * 60)

if not executorch_disponible:
    print("\n⚠️ ExecuTorch no se pudo instalar.")
    print("\n📝 ANÁLISIS TÉCNICO:")
    print("ExecuTorch está en desarrollo activo y aún no tiene")
    print("compatibilidad nativa con todos los frameworks de visión")
    print("por computador. Como alternativas industriales probadas:")
    print("  ✅ ONNX (usado por Microsoft, Facebook, Amazon)")
    print("  ✅ TorchScript (optimizado para mobile)")

else:
    try:
        print("\n🔄 Intentando exportar el modelo con ExecuTorch...")

        import torch
        from executorch.exir import to_edge

        # Cargar el modelo en formato torch
        modelo_torch = torch.load('/content/best.pt', map_location='cpu', weights_only=False)

        if isinstance(modelo_torch, dict):
            # Si es un dict (caso Ultralytics), extraemos el modelo
            modelo_pytorch = modelo_torch['model'].float()
        else:
            modelo_pytorch = modelo_torch

        modelo_pytorch.eval()

        # Input de ejemplo
        example_input = (torch.randn(1, 3, 640, 640),)

        # Exportar
        print("🔄 Convirtiendo a formato ExecuTorch...")
        exported = torch.export.export(modelo_pytorch, example_input)
        edge_program = to_edge(exported)
        executorch_program = edge_program.to_executorch()

        # Guardar
        et_path = '/content/best.pte'
        with open(et_path, 'wb') as f:
            f.write(executorch_program.buffer)

        import os
        size_mb = os.path.getsize(et_path) / 1024 / 1024
        print(f"\n✅ Exportación a ExecuTorch exitosa!")
        print(f"📁 Modelo guardado en: {et_path}")
        print(f"📊 Tamaño: {size_mb:.2f} MB")

    except ImportError as e:
        print(f"\n⚠️ Módulo no disponible: {e}")
        print("\n📝 ExecuTorch requiere dependencias adicionales que no")
        print("están disponibles en el entorno de Colab estándar.")

    except Exception as e:
        print(f"\n⚠️ Error en exportación: {type(e).__name__}")
        print(f"   Mensaje: {str(e)[:300]}")
        print("\n📝 ANÁLISIS DEL ERROR:")
        print("YOLOv8 utiliza operadores específicos (como NMS y")
        print("postprocesamiento dinámico) que aún no son totalmente")
        print("compatibles con el grafo estático que requiere ExecuTorch.")
        print("\n💡 Esto es un problema CONOCIDO en la comunidad y se")
        print("está trabajando activamente en su solución.")

print("\n" + "=" * 60)
print("✅ Proceso de exportación finalizado")
print("=" * 60)
```
